##  estimator median (calculate  median is the middle value of a dataset

## Saturation means color intensity

```python
# import Libararies
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt


#  load Data set
flower1= sns.load_dataset("titanic")
flower1

# Draw a bar plot
sns.barplot(x="class" , y="fare" , hue= "sex" , data=flower1 , estimator=median, saturation= 0.2)
           
plt.show()
```

```python
# import Libararies
import seaborn as sns
import numpy 
import matplotlib.pyplot as plt


#  load Data set
flower1= sns.load_dataset("titanic")
flower1

# Draw a bar plot
sns.barplot(x="class" , y="fare" , data=flower1 , 
                 linewidth=4.5, errcolor=".2" ,  # error bar color,  linewidth mean outline width of bars
                 color=(1, 1, 1, 0) , edgecolor=".2")  # edge color of bars , color = tranparent bar fills
                     
plt.show()
```

```python
# Import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
flower1 = sns.load_dataset("titanic")

# Draw a bar plot
sns.barplot( x="class",  y="fare", data=flower1,
    linewidth=3.5,          # Outline width of bars
    errcolor=".2",          # Error bar color
    edgecolor=".2",         # Edge color of bars
    color=(0.1, 0.3, 0.3, 0.5)      # Transparent bar fill
)

# Show plot
plt.show()
```

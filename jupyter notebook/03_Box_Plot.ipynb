```python

import matplotlib.pyplot as plt
import seaborn as sns
kashti = sns.load_dataset("titanic")
sns.boxplot(x = "survived" , y= "age" , data=kashti)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

p1 = sns.boxplot(x = "survived" , y= "age" , showmeans= True ,
                 meanprops={ "marker": "*",
                           "markersize": "15" ,
                           "markeredgecolor" : "red"},  data=kashti)
# show labels
plt.xlabel("How many survived" , size = 16 , weight = 'bold')
plt.ylabel("Age (years)" , size = 16 , weight = 'bold')
plt.title("Box Plot" , size = 20 , weight = 'bold')


plt.show() 
```
